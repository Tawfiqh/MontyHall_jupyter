```python
import random

def println(x):
    if logging: print(x, end='')


def setup_monty_hall():
    result = ["goat", "goat", "goat"]
    i = random.randint(0, 2)
    result[i] = "car"
    return result;

def printDoor(pick):
    if logging == False: return
    translate = ["A", "B", "C"];
    println(translate[pick])
    println("  ")

def opensDoor(montyHall, currentPick):
    doors = [0,1,2];
    doors.remove(currentPick);

    # If it's a car randomly open one of the other two goats.
    if montyHall[currentPick] == "car":
        openDoor = random.randint(0, 1)
        openedDoor = doors[openDoor]

    else: #It's a goat so open the other goat.
        doors.remove(montyHall.index("car"))
        # Only thing left in doors now is neither pick nor car.
        # ----- This is the crux of the MontyHall Problem -----
        openedDoor = doors[0]

    return openedDoor;

def montyHallGame(rounds, change, changeTwoRandom = False):
    wins = 0;

    for x in range(rounds):
        montyHall = setup_monty_hall()

        # Pick number one - the first door.
        chosenDoor = random.randint(0, 2)
        # println("1: ")
        printDoor(chosenDoor)

        # opens one door
        openedDoor = opensDoor(montyHall, chosenDoor)
        println("( üêê: ")
        printDoor(openedDoor)
        println(")  ");

        # Now the game show contestant is offered if they want to switch their pick.
        
        # They randomly pick one of the closed door.
        # (Maybe they re-pick their existing door, or they pick the other unopened door)
        if changeTwoRandom:
            doors= [0,1,2]
            doors.remove(openedDoor);
            newPick = random.randint(0, 1)
            chosenDoor = doors[newPick]

        # They switch to the other door
        elif change:
            doors= [0,1,2]
            doors.remove(chosenDoor)
            doors.remove(openedDoor);
            chosenDoor = doors[0]# Now the only door left in the doors array is the one not picked or open.
        # println("2: ") # + str(chosenDoor))
        printDoor(chosenDoor)

        result = montyHall[chosenDoor]
        println(result);

        if(result == "car"):
            wins += 1

        if logging: print()

    print(wins)
    print(str(round(wins / rounds * 100, 2)) + "%")

    return wins;


def interactiveGame(verbose, rounds):
    global logging
    logging = verbose
    random.seed()

    print("")
    print("No of Wins - No change of selection - stick with original door")
    montyHallGame(rounds, False)


    print("")
    print("No of Wins - Always change selected door when offered")
    montyHallGame(rounds, True)


    print("")
    print("No of Wins - Change selected door at random when offered to change")
    montyHallGame(rounds, True, True)

    

from ipywidgets import interact, interactive, fixed, interact_manual

interact(interactiveGame, verbose=False, rounds=(0, 10000));
```
